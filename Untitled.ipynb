## Building the circuit

```python
# Create a Quantum Circuit acting on a quantum register of three qubits

circ = qk.QuantumCircuit(3)
```

```python
# Add a Hadamard Gate on qubit 0, putting this qubit in superposition.

circ.h(0)
```

### Visualize Circuit

```python
circ.draw()
```

```python
# Add a CX (CNOT) Gate on control qubit 0 and target qubit 1,
# putting the qubits in a Bell State.

circ.cx(0, 1)

# Add a CX Gate on contrl qubit 0 and target qubit 2,
# putting the qubits in GHZ state. (|000> + |111>)/2^1/2

circ.cx(0, 2)
```

```python
circ.draw()
```

```python
circ.draw('mpl')
```

```python
# the index of the q, is the exponent of 2. 
# for example, q0 -> 2^0, q1 -> 2^1, q2 -> 2^2

# q0 is the least significant bit(LSB), and q2 is the most significant bit(MSB).
# so, if q0 = 1, q1 = 0, q2 = 0 
# the state is written as  |   0   0   1   >
#                             2^2 2^1 2^0
```

```python
# if you want to save the circuit diagram,

image = circ.draw('mpl')
image.savefig('sample circuit')
```

## Simulating Circuits using Qiskit AER

### Statevector Backend

Statevector simulator -> Ideal result

It is different from the quantum simulator.

```python
# import Aer
import qiskit_aer as qkaer
from qiskit_aer import Aer

# Run the quantum circuit on a statevector simulator backend
statevec_sim_backend = Aer.get_backend('statevector_simulator')
```

```python
job = statevec_sim_backend.run(circ) # execute function is deprecated
```

```python
job.status()
```

```python
result = job.result()
```

```python
outputstate = result.get_statevector(decimals=3)
print(outputstate)
```

```python
from qiskit.visualization import plot_histogram

plot_histogram(result.get_counts())
```

### Unitary Backend

입력 값이나 결과 값에 상관없이, 그냥 어떤 Unitary Operation이 적용됐는지 보고 싶을 때 사용.

```python
# Run the quantum circuit on a unitary simulator backend
unitary_sim_backend = Aer.get_backend('unitary_simulator')
job = unitary_sim_backend.run(circ)
result = job.result()

# Show the result
print(result.get_unitary(decimals=3)) # Unitary Matrix 출력, qubit 순서 주의
```

### OpenQASM Backend

Hadamard Gate 등 Unitary Operator들이 완벽하게 구현되었다고 가정하고 simulation.

확률적으로 |000>과 |111>이 50:50으로 나오지 않으니까, 그거에 대한 simulation.

```python
# 측정 결과의 저장을 위한 classical bit도 필요.
# Create a Quantum Circuit
meas = qk.QuantumCircuit(3, 3) # (qubit #, classical bit #)
meas.barrier([0, 1, 2]) # or meas.barrier(range(3))

# Map the quantum measurement to the classical bits
meas.measure([0, 1, 2], [0, 1, 2]) # or meas.mearsure(range(3), range(3
# measurement of qubit[0] -> classicalBit[0]
# measurement of qubit[1] -> classicalBit[1]
# measurement of qubit[2] -> classicalBit[2]

# The Qiskit circuit objeect supports composition using the addition operator.
qc = circ & meas # '+' operator is deprecated

# Draw the circuit
qc.draw('mpl')
```

```python
# Use Aer's qasm_simulator
qasm_sim_backend = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator.
# Set the number of repeats of the circuit to be 1024, which is default.
job = qasm_sim_backend.run(qc, shots=1024)

# Grab the results from the job.
result_sim = job.result()
```

```python
counts = result_sim.get_counts(qc)
print(counts)
```

```python
from qiskit.visualization import plot_histogram

plot_histogram(counts)
```

```python
# 측정 결과의 저장을 위한 classical bit도 필요.
# Create a Quantum Circuit
meas = qk.QuantumCircuit(3, 3) # (qubit #, classical bit #)
meas.barrier([0, 1, 2]) # or meas.barrier(range(3))

# Map the quantum measurement to the classical bits
meas.measure([0, 1, 2], [0, 1, 2]) # or meas.mearsure(range(3), range(3
# measurement of qubit[0] -> classicalBit[0]
# measurement of qubit[1] -> classicalBit[1]
# measurement of qubit[2] -> classicalBit[2]

# The Qiskit circuit objeect supports composition using the addition operator.
qc = circ & meas # '+' operator is deprecated

# Draw the circuit
qc.draw('mpl')
```

```python
from qiskit import transpile

# Transpile the ideal circuit to a circuit that can be directly executed by the backend
transpiled_circuit = transpile(qc, backend)
transpiled_circuit.draw('mpl', style="iqp")

# Run the transpiled circuit using the simulated fake backend
sampler = SamplerV2(backend)
job = sampler.run([transpiled_circuit])
pub_result = job.result()[0]
counts = pub_result.data.c.get_counts()
plot_histogram(counts)
```
